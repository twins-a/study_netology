### Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

Примеры работы:

Введите команду:

2.
Введите команду:

Введите номер документа:

Результат:

```python
def document_owner(documents, number_doc = -1):
    """
    Функция для определения владельца документа по номеру документа
    
    На вход подается список из словарей со структурой: 
    {'type': 'xxx', 'number': 'yyyyyy', 'name': 'zzzzzzzzzzzz'}
    
    Возвращает:
                имя владельца: если номер найден 
            или
                None: если документ в базе не найден
    
    """
    if number_doc == -1:
        number_doc = input('Введите номер документа: ')
        
    for i in documents:
        if i['number'] == number_doc:
            return i['name']
    return
    
    
# document_owner(documents)
# print(document_owner(documents))
# ?document_owner

n_doc = document_owner(documents)
if n_doc is None:
    print(f"Документ не найден в базе")
else:
    print(f"Владелец документа: {n_doc}")
```

### Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится

Примеры работы:
1.
Введите команду:
s

Введите номер документа:
10006
Результат:
Документ хранится на полке: 2

2.
Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе

```python
# считаю что если документ не найден на полке, то такого документа нет в базе (то есть нет в documents)
def document_directories(directories, number_doc = -1):
    """
    Функция для определения на какой полке хранится документ
    
    Возвращает:
                номер полки на которой находится документ
            или
                0 если документ не найден в базе

    """
    if number_doc == -1:
        number_doc = input('Введите номер документа: ')
    
    for i, shelf in enumerate(directories.values()):
        if number_doc in shelf:
            return i+1
    return 0
    
n_shelf = document_directories(directories)
if n_shelf > 0:
    print(f'Документ хранится на полке: {n_shelf}')
else:
    print(f'Документ не найден в базе')
```

```python
def dict_keys_to_list(dict_keyss):
    """
    Преобразование словаря ключей (списка) в строку
    
    """
    return ", ".join(dict_keyss)

dict_keys_to_list((['1', '2', '3', '10']))
```

```python
def add_shelf(directories):
    """
    Функция для добавления новой полки
    
    """
    number_shelf = input('Введите номер полки: ')
    ald_length_shelf = len(directories)
    directories.setdefault(number_shelf)
    if len(directories) > ald_length_shelf:
        print(f'Полка добавлена. Текущий перечень полок: {dict_keys_to_list(directories.keys())}.')
    else:
        print(f'Такая полка уже существует. Текущий перечень полок: {dict_keys_to_list(directories.keys())}.')
    
    
add_shelf(directories)
```

```python
def del_document(documents, directories):
    """
    Функция для удаления документа из базы
    
    
    """
    num_doc = input('Введите номер документа: ')
    
#    num_doc = '42'
#    num_doc = '11-2'
#    num_doc = '10006'
#    num_doc = '423'
#    num_doc = '2207 876234'
    
    number_dir = document_directories(directories, num_doc)
    if number_dir > 0:
        # удаляем документ с полки:
        number_dir_str = str(number_dir)
        directories[number_dir_str].remove(num_doc)
      
        # удаляем сам документ:
        del(documents[number_dir])
        
        print('Документ удален.')
    
    else:
        print(f'Документ не найден в базе')

    print('\nТекущий список документов:')
    list_document(directories)

    
    
del_document(documents, directories)
```

```python
def move_document(directories):
    """
    Функция для переноса документа из базы
    
    
    """
        
    num_doc = input('Введите номер документа: ')

#    num_doc = '42'
#    num_doc = '11-2'
#    num_doc = '10006'
#    num_doc = '423'
#    num_doc = '2207 876234'


    number_dir = document_directories(directories, num_doc)
    if number_dir > 0:
        num_dir_new = input('Введите номер папки: ')
        # проверим есть ли такая папка:
        if num_dir_new in directories:
            # удаляем документ со старой полки:
            directories[str(number_dir)].remove(num_doc)
            
            # добавляем документ на новую полку:
            directories[num_dir_new].append(num_doc)
            print('Документ перемещен. ')
            
        else:
            print('Такой полки не существует.')
        
    else:
        print(f'Документ не найден в базе')
   
    print('\nТекущий список документов:')
    list_document(directories)
    
move_document(directories)
```

```python
def main(documents):
    while True:
        command_user = input('Введите команду:\n   p - владелец по номеру документа;\n'
                             +'   s - полка на которой хранится документ;\n   l - полный список документов;\n'
                             +'   ads - добавить полку; \n   ds - удалить полку;\n   ----------------------\n'
                             +'   ad - добавить новый документ;\n   d - удалить документ;\n'
                             +'   m - переместить документ с полки на полку;\n'
                             +'   ----------------------\n   q-завершить программу\n\n => ')
        print()
        # Определение владельца документа по номеру документа
        if command_user == 'p':
            #document_owner(documents)
            doc_owner = document_owner(documents)
            if doc_owner is None:
                print(f"Документ не найден в базе")
            else:
                print(f"Владелец документа: {doc_owner}")
            
        # Определение на какой полке хранится документ
        elif command_user == 's':
            print(document_directories(directories))
            
        # Вывод полной информации по документам
        elif command_user == 'l':
            list_document(documents, directories)
            
        # добавление новой полки
        elif command_user == 'ads':
            add_shelf(directories)
            
        # удаление полки (если не пустая)
        elif command_user == 'ds':
            del_shelf(directories)
            
        # -------------------------------
        # добавление документа
        elif command_user == 'ad':
            add_document(documents, directories)
            
        # удаление документа из базы
        elif command_user == 'd':
            del_document(documents, directories)
            
        # Перемещение документа с полки на полку
        elif command_user == 'm':
            move_document(directories)
            
        # -------------------------------
        
        # Завершение программы
        elif command_user.lower() == 'q':
            print('До свидания!')
            break
            
        # Проверка на незнакомые команды
        else:
            print('Данная команда не предусмотренна ')
        
        print()

main(documents)
```
