### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
•	если источник traffic_source равен yandex или google, то в source_type ставится organic
•	для источников paid и email из России - ставим ad
•	для источников paid и email не из России - ставим other
•	все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd
```

```python
def source_type_f(par_row):
    """
    Определение поля source_type
    На вход подается строка из dataframe
    На выходе строка-значение для поля source_type по условию выше.
    """
    if (par_row.traffic_source in ['yandex', 'google']):
        return 'organic'
    elif (par_row.traffic_source in ['paid', 'email']) and (par_row.region == 'Russia'):
         return 'ad'
    elif (par_row.traffic_source in ['paid', 'email']) and (par_row.region != 'Russia'):
         return 'other'
    else:
        return par_row.traffic_source
```

```python
# Вариант 1. через функцию:

logs = pd.read_csv('Files_Pandas1/visit_log.csv', sep=';')
logs['source_type'] = logs.apply(source_type_f, axis=1)
logs.head(10)
```

```python
# Вариант 2. через pd.loc()

logs = pd.read_csv('Files_Pandas1/visit_log.csv', sep=';')
logs.loc[logs.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'
logs.loc[(logs.traffic_source.isin(['paid', 'email'])) & (logs.region == 'Russia'), 'source_type'] = 'ad'
logs.loc[(logs.traffic_source.isin(['paid', 'email'])) & (logs.region != 'Russia'), 'source_type'] = 'other'
logs.loc[~logs.traffic_source.isin(['yandex', 'google', 'paid', 'email']), 'source_type'] = logs.traffic_source
logs.head(10)
```

```python
# Вариант 3. через pd.loc() Заменил последнее условие...

logs = pd.read_csv('Files_Pandas1/visit_log.csv', sep=';')
logs.loc[logs.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'
logs.loc[(logs.traffic_source.isin(['paid', 'email'])) & (logs.region == 'Russia'), 'source_type'] = 'ad'
logs.loc[(logs.traffic_source.isin(['paid', 'email'])) & (logs.region != 'Russia'), 'source_type'] = 'other'
logs.loc[pd.isnull(logs.source_type), 'source_type'] = logs.traffic_source
logs.head(10)
```

1. Используйте файл с оценками фильмов ml-latest-small/ratings.csv.
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('Files_Pandas1/ratings.csv')
ratings
```

```python
# Получим пользователей которые выставили более 100 оценок
users_100 = ratings.groupby('userId').count().reset_index()
users_100 = users_100[users_100.rating > 100]
users_100 = users_100[['userId', 'movieId']]
users_100.columns = ['userId', 'quantity']
users_100
```

```python
# Получим среднее время жизни для всех пользователей в секундах
delta = ratings.groupby('userId').agg({'timestamp': [min, max]}).reset_index()
delta.loc[:, 'lifetime'] = delta['timestamp']['max'] - delta['timestamp']['min']
delta = delta[['userId', 'lifetime']]
delta.columns = ['userId', 'lifetime']
delta
```

```python
# теперь объединим два dataframe

users_100.merge(delta).sort_values('lifetime', ascending=False)
```

```python
# Очень смущает время жизни по некоторым строкам. 
# например: 205 строка, userId=536. выставлено 109 оценок, а время жизни всего 206 сек. = около 3,5 миут
```

```python
# Получим среднее время жизни для всех пользователей в секундах 
# И второй столбец для проверки разница timestamp переведенная в дату
delta = ratings.groupby('userId').agg({'timestamp': [min, max]}).reset_index()
delta.loc[:, 'lifetime'] = delta['timestamp']['max'] - delta['timestamp']['min']
delta.loc[:, 'lifetime1'] = pd.to_datetime(delta['timestamp']['max'].astype(int), unit='s') - pd.to_datetime(delta['timestamp']['min'].astype(int), unit='s')
delta = delta[['userId', 'lifetime', 'lifetime1']]
delta.columns = ['userId', 'lifetime', 'lifetime1']
delta
```

```python
# теперь объединим два dataframe в кот. два столбика с разницей дат (время жизни)

users_100.merge(delta).sort_values('lifetime', ascending=False)
```

### Задание 4

2. Дана статистика услуг перевозок клиентов компании по типам
(см. файл “Python_13_join.ipynb” в разделе Материалы для лекции «Продвинутый pandas» - Ноутбуки к лекции «Продвинутый pandas»).

Необходимо сформировать две таблицы:
• таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
• аналогичную таблицу по типам выручки с указанием адреса клиента


Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки
auto - автомобильные перевозки
air - воздушные перевозки
client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# таблица 1. Без адресов

table1 = rzd.merge(auto, how='outer')
table1 = table1.merge(air, how='outer')
table1
# значения NaN не меняю на 0 специально.
```

```python
# таблица 2. С адресами

table2 = table1.merge(client_base, how='outer')
table2
```
